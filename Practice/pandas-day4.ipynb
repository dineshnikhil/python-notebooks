```python
df_jobs = pd.read_csv('jobs.csv')
df_jobs
```

```python
df_jobs.groupby("Agency")[['Salary Range From', 'Salary Range To']].mean().head(5)
```

```python
def rate_a_agency(count):
    if count < 70:
        return 'above average'
    elif count > 70 and count < 300:
        return 'Good'
    else:
        return 'excelent'

df_jobs.groupby('Agency').count()['Job ID'].apply(rate_a_agency)
```
