```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = x*1.5
y
```

```python
plt.plot(x,y)
plt.xlabel('x lable')
plt.ylabel('y lable')
plt.xlim(1, 8)
plt.ylim(1, 14)
plt.title("first ploting")
plt.savefig('myfirstplot.png')
```

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig_1 = plt.figure()

axes_1 = fig_1.add_axes([0,0,1,1])
axes_1.plot(a,b)

axes_1.set_xlim(0, 8)
axes_1.set_ylim(0, 8000)
axes_1.set_xlabel("A lable")
axes_1.set_ylabel("B lable")
axes_1.set_title("Power of 4")

axes_2 = fig_1.add_axes([0.2,0.4,0.5,0.5])
axes_2.plot(x,y)

axes_2.set_xlim(0, 8)
axes_2.set_ylim(0, 16)
axes_2.set_xlabel("X lable")
axes_2.set_ylabel("Y lable")
axes_2.set_title("power of 2")
```

```python
# Set the mean and standard deviation
mu = 1
sigma = 2

# Generate x values from -3 to 3 with a step size of 0.1
x = np.arange(-3, 3, 0.1)

# Calculate y values using the normal distribution function
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x, y)
```

```python
# Set the constants
a = 0.1
b = 0.2

# Generate theta values from 0 to 10pi with a step size of 0.1
theta = np.arange(0, 10*np.pi, 0.1)

# Calculate r values using the logarithmic spiral equation
r = a * np.exp(b * theta)

# Convert polar coordinates to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plot the spiral
plt.plot(x, y)
plt.axis('equal')
```
