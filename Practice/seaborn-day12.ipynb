```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/dm_office_sales.csv')
df.head()
```

```python
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
# in hue we can choose the categorical column like level of education
sns.scatterplot(x='salary', y='sales', hue='level of education', data=df)
```

```python
# in hue we can choose the continuous column like level of education
sns.scatterplot(x='salary', y='sales', hue='salary', data=df)
```

```python
# in seaborn we can scatter the plot size based the column
sns.scatterplot(x='salary', y='sales', data=df, size='work experience')
```

```python
sns.scatterplot(x='salary', y='sales', data=df, s=200, alpha=0.2)
```

```python
sns.scatterplot(x='salary', y='sales', hue='level of education', data=df, style='level of education')
```

```python
fig = plt.figure(figsize=(15, 5))
fig.set_facecolor('#C69B7B')
chart = sns.scatterplot(x='salary', y='sales', hue='level of education', data=df, style='level of education')
chart.set_facecolor('#3A3845')
```

```python
sns.displot(data=df, x='salary')
```

```python
sns.displot(data=df, x='salary', kde=True)
```

```python
sns.kdeplot(data=df, x='salary', fill=True)
```
