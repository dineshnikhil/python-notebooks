##### 7.Create a new column called 'FamilySize' that contains the sum of the 'SibSp' and 'Parch' columns for each passenger.

##### 8.Create a new column called 'AgeCategory' that categorizes passengers into "Child" (age < 18), "Adult" (age >= 18 and age < 65), and "Senior" (age >= 65) based on their age.

```python
def define_age_category(age):
    if (age < 18):
        return "Child"
    elif (age >= 18 and age < 65):
        return "Adult"
    else:
        return "Senior"
    
titanic['AgeCategory'] = titanic['Age'].apply(define_age_category)
titanic
```
