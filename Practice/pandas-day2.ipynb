```python
def last_four_digits(num):
    return str(num)[-3:]

last_four_digits(123456789)
```

```python
df['last_four_num'] = df['CC Number'].apply(last_four_digits)
df.head()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price > 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
df.head(20)
```

```python
def quality(total_bill, tip):
    if tip / total_bill > 0.25:
        return 'Generous'
    else:
        return 'other'
```

```python
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df : quality(df['total_bill'], df['tip']), axis=1)
```

```python
df['tip_quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df[(df['tip'] /df['total_bill']) > 0.25].head(5)
```
