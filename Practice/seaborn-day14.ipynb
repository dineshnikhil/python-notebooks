### Matrix plots

```python
df_1 = pd.read_csv('./data/country_table.csv')
df_1
```

```python
# to use heatmap we should have nameing index
df_1 = df_1.set_index('Countries')
df_1
```

```python
sns.heatmap(df_1)
```

```python
sns.heatmap(df_1.drop('Life expectancy', axis=1), linewidths=.2, annot=True)
```

```python
sns.heatmap(df_1.drop('Life expectancy', axis=1), linewidths=.2, annot=True, cmap='viridis')
```

```python
sns.heatmap(df_1.drop('Life expectancy', axis=1), linewidths=.2, annot=True, cmap='viridis', center=40)
```

```python
sns.clustermap(df_1.drop('Life expectancy', axis=1), linewidths=.2, annot=True, cmap='viridis')
```
