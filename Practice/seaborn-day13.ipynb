```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/dm_office_sales.csv')
df.head()
```

```python
sns.scatterplot(x='salary', y='sales', hue='work experience', data=df)
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='division', hue='level of education')
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd')
```

```python
df.head()
```

```python
df.groupby('level of education')['salary'].agg(['mean', 'std'])
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data=df, x='level of education', y='salary', hue='division', estimator=np.mean, errorbar='sd')

plt.legend(bbox_to_anchor=(1.05, 1))
```

### Grid Plots

```python
student_df.head()
```

```python
sns.catplot(data=student_df, x='gender', y='math score', kind='box')
```

```python
sns.catplot(data=student_df, x='gender', y='math score', kind='box',  col='test preparation course')
```

```python
sns.catplot(data=student_df, x='gender', y='math score', kind='box', row='lunch', col='test preparation course')
```

```python
gd = sns.PairGrid(student_df)

gd.map_upper(sns.scatterplot)
gd.map_diag(sns.kdeplot)
gd.map_lower(sns.kdeplot)
```

```python
gd = sns.PairGrid(student_df, hue='gender')

gd.map_upper(sns.scatterplot)
gd.map_diag(sns.kdeplot, fill=True)
gd.map_lower(sns.kdeplot, fill=True)
```
