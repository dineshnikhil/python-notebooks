```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.arange(1, 11)

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x**2, label='x**2', color='#FFD3B0', lw=3)
axes.plot(x, x**3, label='x**3', color="#FF6969", lw=3)
axes.plot(x, x**4, label='x**4', color="#FFD93D", lw=3)

axes.legend()
```
