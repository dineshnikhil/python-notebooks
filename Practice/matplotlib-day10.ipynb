### Matplotlib Day10

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set the mean and standard deviation
mu = 1
sigma = 2

# Generate x values from -3 to 3 with a step size of 0.1
x = np.arange(-3, 3, 0.1)

# Calculate y values using the normal distribution function
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

fig = plt.figure()
# bigger plot
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y)

# zommed version of above plot
axes1 = fig.add_axes([0.4, 0.1, 0.5, 0.5])
axes1.plot(x,y)
axes1.set_xlim(-1, 3)
axes1.set_ylim(0.125, 0.200)
```

```python
a = np.arange(0,10)
b = a * 2
fig, axes = plt.subplots(1,2)
fig.set_figwidth(10)

axes[0].plot(x,y)
axes[1].plot(a, b)
```

```python
# Set the constants
a_1 = 0.1
b_2 = 0.2
# Generate theta values from 0 to 10pi with a step size of 0.1
theta = np.arange(0, 10*np.pi, 0.1)
# Calculate r values using the logarithmic spiral equation
r = a_1 * np.exp(b_2 * theta)
# Convert polar coordinates to Cartesian coordinates
c = r * np.cos(theta)
d = r * np.sin(theta)

e = np.linspace(0, 10, 11)
f = e ** 4

```

```python
fig, axes = plt.subplots(2,2)
fig.set_figwidth(12)
fig.set_figheight(10)
fig.suptitle("Differnt Types of Graphs")
fig.subplots_adjust(wspace=0.3, hspace=0.2)

axes[0][0].plot(a,b)
axes[0][0].set_title("Linear Graph")

axes[0][1].plot(e,f)
axes[0][1].set_title("Exponentional Graph")

axes[1][0].plot(x,y)
axes[1][0].set_title("Bell Curve Graph")

axes[1][1].plot(c,d)
axes[1][1].set_title("Spiral Graph")
```
